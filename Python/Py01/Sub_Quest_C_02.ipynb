## Q1-1. 회문 판독기

문자열을 입력으로 받으면 해당 문자열이 회문인지 아닌지 판독하여 회문인 경우 True, 아닌 경우 False를 반환하는 함수를 만듭니다.

이때, 반환되는 True or False는 문자열이 아닌 'bool' 타입입니다.

---
입력 예시:
```python
target1 = "아많다많다많다많아"
target2 = "아정말많다많다많아"

print(is_palindrome(target1))
print(is_palindrome(target2))
```

출력 예시:
```
True
False
```

````python
def is_palindrome(text):
    # [[YOUR CODE]]
    reversed_text = ''
    text2 = text
    while text2:
      reversed_text += text2[-1]
      text2 = text2[:-1]

    # print(reversed_text)
    if text == reversed_text:
      return True

    return False
````

````python
# Test
target1 = "여보게저기저게보여"
target2 = "보이긴뭐가보여저기"

print(is_palindrome(target1))
print(is_palindrome(target2))
````

## Q1-2. 회문 분류기

Q1-1에서 만든 판독기 함수를 이용하여 회문 분류기 함수를 만들어 봅니다.

회문 분류기 함수는 아래의 작업을 수행합니다.

1. 여러 단어가 포함된 리스트를 입력으로 받으면 한 단어씩 판독기를 통해 회문 여부를 확인 후 하나의 딕셔너리에 회문과 회문이 아닌 단어를 분류하여 저장합니다.
    - 이때, 회문의 경우 딕셔너리에서 key값은 "회문O", value값은 회문인 단어들로 이루어진 리스트입니다.
    - 회문이 아닌 경우의 key값은 "회문X", value값은 회문이 아닌 단어들로 이루어진 리스트입니다.

2. 모든 단어의 분류가 완료되면 저장한 딕셔너리와, 회문의 총 개수, 전체에서 회문의 비율을 계산하여 반환합니다.
    - 이때, 회문의 총 개수는 int, 회문의 비율은 float입니다.
    - 회문의 비율이 `30%`라면 반환되는 값은 `0.3`입니다.

---
예시 입력:
```python
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

print(palindrome_classifier(targets))
```

예시 출력:
```
({'회문O': ['기러기', '토마토', '스위스', '인도인', '별똥별', '우영우', '우향우'], '회문X': ['까마귀', '스윗소스', '한국인']}, 7, 0.7)
```

````python
def palindrome_classifier(targets):
    # [[YOUR CODE]]
    palindromes, not_palindromes = [], []
    for t in targets:
        if is_palindrome(t):
            palindromes.append(t)
        else:
            not_palindromes.append(t)
    result = { '회문O': palindromes, '회문X': not_palindromes }
    count_palindromes = len(result['회문O'])
    ratio_palindromes = 1 - (1 / len(targets) * len(not_palindromes))

    return result, count_palindromes, ratio_palindromes

````

````python
# Test
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

print(palindrome_classifier(targets))
````

## Q2-2. 성적 분석기

이제 Q2-1에서 만든 변수를 참조하여 학생의 이름을 인자로 받아 아래의 예시와 같이 성적을 분석해주는 함수를 만들어 봅니다.

---

예시 입력:
```python
score_analysis("설지민")
```
예시 출력:
```
[설지민 성적 분석]
국어: 88점, 학급 평균(91.50점) 대비 -3.50
영어: 96점, 학급 평균(87.75점) 대비 +8.25
수학: 90점, 학급 평균(87.00점) 대비 +3.00

전과목 평균: 91.33점, 학급 평균(88.75점) 대비 +2.58
```


````python
def score_analysis(name):
    # [[YOUR CODE]]
    print(f'[{name} 성적 분석]')
    class_averages = []

    for key, value in scores[name].items():

        # 학급 평균 구하기
        class_scores = [score[key] for sub, score in scores.items()]
        class_average = sum(class_scores) / len(class_scores)

        # 대비 구하기
        score_diff = value - class_average

        print(f'{key}: {value}점, 학급 평균({class_average:.2f}점) 대비 {"" if score_diff < 0 else "+"}{score_diff:.2f}')

        class_averages.append(class_average)

    # 전과목
    # all_values = [v for v in scores.values()]
    print(list(zip(*scores.values())))
    # print(all_values)
    # print()
    # all_scores = [v.values() for v in all_values]
    # # all_scores = list(zip(*all_values))
    # print(all_scores)

    # all_scores = sum(all_scores) / len(all_scores)
    # all_sub_average = sum([all_scores])
    # class_average = sum(class_averages) / len(class_averages)

    # print()
    # print(f'전과목 평균: {total_average}점, 학급 평균({class_average:.2f}점) 대비 {"" if score_diff < 0 else "+"}{score_diff:.2f}')
````
